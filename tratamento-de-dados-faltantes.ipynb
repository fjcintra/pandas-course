## Tratamento de dados faltantes com pandas.dropna()

```python
import pandas as pd
```

```python
dados = pd.read_csv('dataset/aluguel_residencial.csv', sep=';')
dados.head(10)
```

## Tratamento de dados faltantes com pandas.fillna()

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

```python
a = dados.shape[0]
dados = dados[~selecao]
b = dados.shape[0]
a- b
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
dados.fillna(0, inplace=True)
```

```python
dados
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
dados.info()
```

```python
dados.to_csv('dataset/aluguel_residencial.csv', sep=';', index=False)
```
