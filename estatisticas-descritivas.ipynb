# Criando agrupamentos

```python
import pandas as pd
dados = pd.read_csv('dataset/aluguel_residencial.csv', sep = ';')
dados.head(10)
```

```python
# bairros = dados['Bairro'].unique()
# bairros
```

```python
grupo_bairros = dados.groupby('Bairro')
grupo_bairros['Valor', 'Condominio'].mean().round(2)
```

# Estatístca descritiva

```python
grupo_bairros['Valor'].describe().round(2)
```

```python
grupo_bairros['Valor'].aggregate(['min', 'max', 'sum'])
```

```python
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (30,5))
```

```python
grupo_bairros['Valor'].std().plot.bar(color = 'blue')
```

```python
grupo_bairros['Valor'].mean().plot.bar(color = 'blue')
plt.ylabel('Valor do Aluguel')
plt.title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
```
